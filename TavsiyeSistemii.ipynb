# Fİlm Tavsiye Sistemi 





```python
import numpy as np
import pandas as pd
```

```python
#Veri setini okuyoruz ve sütunları isimlendiriyoruz. Kullanıcıların filmlere verdiklerri oyları tutan bir veri seti.
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('users.data', sep='\t', names=column_names)
```

```python
df.head() 
# ilk 5 veriyi okuyoruz
```

```python
# Kaç kayıt var:
len(df)
```

### Filmlerin ve id'lerinin olduğu veri setini okuyoruz.

```python
#Diğer veri setindeki item_id buradan geliyor iki dosyayı veritabanlarındaki kullanıma benzer şekilde kullanıyoruz.
movies = pd.read_csv("movie_id_titles.csv")
movies.head()
```

```python
#Bu veri setindeki kayıt sayısı:
len(movies)
```

```python
#Bu iki  veri setini birleştiriyoruz, sonucun görünüşü:
df = pd.merge(df, movies, on='item_id')
df.head()
```

### Tavsiye Sistemimizi Kuruyoruz:


```python
#Veri setini düzeneyerek yeni bir veri seti elde ediyoruz, bu veri setinde her satır bir kullanıcı olacak ve her sütun da film
#isimleri olacak ve değerler oy değerleri olacak.Tablonun görünüşü:

movietable = df.pivot_table(index='user_id',columns='title',values='rating')
movietable.head()
```

```python
#bu tablonun tipi DataFrame
type(movietable)
```

### Forrest Gump (1994) filmine benzer film önerileri yapalım.

Forrest Gump (1994) filminin oy oranlarına bakalım:

corrwith() metodunu kullanarak Forrest Gump (1994) filmi ile korelasyonları hesaplatalım:

```python
#Forrest Gump filminin oy oranlarına bakıyoruz.
ratings = movietable["Forrest Gump (1994)"]
ratings.head(10)
```

```python
#Benzerliklerine bakmak için korelasyonlarını hesaplatıyoruz.Bazı kayıtlarda boşluklar olduğu için hata veriyor.
similar= movietable.corrwith(ratings)
```

```python
# Korelasyonun 1 ve -1 arasında olduğunu görüyoruz, 1 uyumlu -1 uyumsuz olduğunu gösteriyor.
similar
```

```python
#Series ile çalıştığımız için hata verdi.
type(similar)
```

#### Dataframe'e dönüştürelim ve NaN kayıtlarını temizleyip bakalım:

```python
corr_ = pd.DataFrame(similar, columns=['Correlation'])
corr_.dropna(inplace=True)
corr_.head()
```

### Elde ettiğimiz dataframe'i sıralayım ve Forrest Gump'a en yakın tavsiye edeceği filmi görelim:

```python
corr_.sort_values('Correlation',ascending=False).head(10)
```

#### Bazı alakasız sonuçlar çıktı, bunun nedeninin bu filmlerin çok az oy alması . Bu durumu düzeltmek için 100'den az oy alan filmleri eliyoruz. Bu amaçla ratings isimli bir dataframe oluşturuyoruz ve burada her fimin kaç tane oy aldığını tutuyotuz.


timestamp sütununa siliyoruz, ihtiyacımız yok.

```python
df.drop(['timestamp'], axis = 1)
```

```python
# Her filmin ortalama (mean value) rating değerini bulalım . Title sütununa göre gruplayıp, rating sütunun ortalamasını alıyor.
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())

# Büyükten küçüğe sıralayoruz.
ratings.sort_values('rating',ascending=False).head()
```

```python
# Şimdi her filmin aldığı oy sayısını buluyoruz.
ratings['rating_oy_sayisi'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
#En çok oy alan filmleri büyükten küçüğe sıralayalım.
ratings.sort_values('rating_oy_sayisi',ascending=False).head()
```

```python
# corr_ dataframe'ine rating_oy_sayisi sütununu ekleyelim 
```

```python
corr_ = corr_.join(ratings['rating_oy_sayisi'])
corr_.head()
```

```python
#Oy sayısı 100'den büyük olanları ve korelasyonu en yüksek olanları gösterecek.
```

### Sonuç: 

```python
corr_[corr_['rating_oy_sayisi']>100].sort_values('Correlation',ascending=False).head()
```

```python
#En iyi korelasyona sahip olarak ilk başta kendini daha sonra benzer filmleri gösteriyor.
```

```python
#Şimdi bunu bir arayüzle kullanıcının kullanabileceği bir hale getirelim.
```

```python
from tkinter import * 
window = Tk()
window.title("Test")
window.geometry("1000x700")

lb_baslik=Label(text="Film Tavsiye Sistemi",font="SansSerif 18",fg="Red",width=40,bg="white")
lb_baslik.place(x=180,y=10)
lb_baslik=Label(text="Lütfen bir film seçiniz.",font="SansSerif 10")
lb_baslik.place(x=30,y=50)
#filmleri kullanıcının seçebileeği bir listeye aktarıyoruz.
data=movies['title']
list=Listbox(window,height="30",width="45",bg="#dddddd",bd="2px",font="SansSerif",fg="black")
list.place(x=20,y=75)
label=Label()
label.place(x=450,y=110)

a=1
for i in data:
    list.insert(a,i)
    a=a+1


lb1=Label(text="")
lb1.place(x=470,y=125)
lb=Label(text="")
lb.place(x=460,y=165)

#Tavsiye işini yapacak olan metot
def metot():
    eleman=list.curselection()
    if(eleman):
        deger=list.get(eleman)
        lb1.configure(text=deger,font="SansSerif 12")
        label.configure(text="")
        ratings = movietable[deger]
        similar = movietable.corrwith(ratings)
        corr_ = pd.DataFrame(similar, columns=['Correlation'])
        corr_.dropna(inplace=True)
        corr_.sort_values('Correlation',ascending=False)
        df.drop(['timestamp'], axis = 1)
        ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
        ratings.sort_values('rating',ascending=False)
        ratings['Oy_Sayısı'] = pd.DataFrame(df.groupby('title')['rating'].count())
        ratings.sort_values('Oy_Sayısı',ascending=False)
        corr_.sort_values('Correlation',ascending=False)
        corr_ = corr_.join(ratings['Oy_Sayısı'])
        text1=corr_[corr_['Oy_Sayısı']>100].sort_values('Correlation',ascending=False).head()
        text1=text1.rename(columns={'title': 'Film', 'Correlation': 'Korelasyon'})
        lb.configure(text=text1, font="SansSerif 12",bg = "white",justify = RIGHT)
        
               
    else:
        label.configure(text="Lütfen seçim yapınız.")
                
b1=Button(text="Tavsiyeleri Göster",command=metot,bg="#F0681F",fg="white")
b1.place(x=450, y=75)

window.mainloop()
```
